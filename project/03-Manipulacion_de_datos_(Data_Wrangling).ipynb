    - `'order_id'`: número de ID que identifica de manera única cada pedido.

- `products.csv`: cada fila corresponde a un producto único que pueden comprar los clientes.

    - `'product_name'`: nombre del producto.
    - `'aisle_id'`: número ID que identifica de manera única cada categoría de pasillo de víveres.
    - `'department_id'`: número ID que identifica de manera única cada departamento de víveres.
- `order_products.csv`: cada fila corresponde a un artículo pedido en un pedido.
    - `'order_id'`: número de ID que identifica de manera única cada pedido.
    - `'product_id'`: número ID que identifica de manera única cada producto.
    - `'add_to_cart_order'`: el orden secuencial en el que se añadió cada artículo en el carrito.
    - `'reordered'`: 0 si el cliente nunca ha pedido este producto antes, 1 si lo ha pedido.
- `aisles.csv`
    - `'aisle_id'`: número ID que identifica de manera única cada categoría de pasillo de víveres.
    - `'aisle'`: nombre del pasillo.
- `departments.csv`

### [C3] Para cada producto, ¿cuál es la proporción de las veces que se pide y que se vuelve a pedir?

```python
# Calcular el número total de pedidos por producto
total_orders_per_product = order_products['product_id'].value_counts()

# Calcular el número de reordenaciones por producto
reorders_per_product = reordered_products['product_id'].value_counts()

# Calcular la tasa de repetición del pedido
reorder_rate = (reorders_per_product / total_orders_per_product).reset_index()
reorder_rate.columns = ['product_id', 'reorder_rate']

# Unir con los nombres de productos
reorder_rate = reorder_rate.merge(products, on='product_id', how='left')

print(reorder_rate[['product_id', 'reorder_rate', 'product_name']].head(20))

# Crear una gráfica de barras
plt.figure(figsize=(10,6))
plt.barh(reorder_rate['product_name'].head(20), reorder_rate['reorder_rate'].head(20), color='skyblue')
plt.xlabel('Reorder Rate')
plt.ylabel('Product Name')
plt.title('Top 20 Products by Reorder Rate')
plt.gca().invert_yaxis()  # invertir el eje y para que el producto con la mayor tasa de reordenación esté en la parte superior
plt.show()
```

Escribe aquí tus conclusiones

La tasa de reordenación por producto proporciona una visión sobre cuáles productos tienen una alta fidelidad de los clientes. ### [C4] Para cada cliente, ¿qué proporción de sus productos ya los había pedido?

```python
# Filtrar los pedidos reordenados
reordered_orders = order_products[order_products['reordered'] == 1]

# Asegurarse de que 'user_id' está presente en order_products mediante un merge si no está
if 'user_id' not in order_products.columns:
    order_products = order_products.merge(instacart_orders[['order_id', 'user_id']], on='order_id', how='left')

# Calcular el número total de productos pedidos por usuario
total_products_per_user = order_products.groupby('user_id')['product_id'].count()

# Calcular el número de productos reordenados por usuario
reordered_products_per_user = reordered_orders.groupby('user_id')['product_id'].count()

# Calcular la proporción de productos reordenados por usuario
reorder_rate_per_user = (reordered_products_per_user / total_products_per_user).fillna(0).reset_index()
reorder_rate_per_user.columns = ['user_id', 'reorder_rate']

# Mostrar los resultados
print(reorder_rate_per_user.head(20))

# Crear una gráfica de barras
plt.figure(figsize=(10,6))
plt.barh(reorder_rate_per_user['user_id'].astype(str).head(20), reorder_rate_per_user['reorder_rate'].head(20), color='skyblue')
plt.xlabel('Reorder Rate')
plt.ylabel('User ID')
plt.title('Top 20 Users by Reorder Rate')
plt.gca().invert_yaxis()  # invertir el eje y para que el usuario con la mayor tasa de reordenación esté en la parte superior
plt.show()
```
